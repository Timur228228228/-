## Problem 2


Write a function that takes a two-dimensional array and a string as input and returns an array rotated 90 degrees counterclockwise if the string 'left' was passed, and clockwise if the string 'right' was passed.

Example for input: $\begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix}$.\
If the string 'left' is passed, the function should return $\begin{bmatrix} 3 & 6 & 9 \\ 2 & 5 & 8 \\ 1 & 4 & 7 \end{bmatrix}$, and if the string 'right' is passed, the function should return $\begin{bmatrix} 7 & 4 & 1 \\ 8 & 5 & 2 \\ 9 & 6 & 3 \end{bmatrix}$.

Do not use numpy and nested loops. Use list comprehensions and built-in function `zip`.

````python
def rotate_90(matrix, direction):
  if direction == 'right':
    return [list(row) for row in zip(*matrix[::-1])]
  if direction == 'left':
      return [list(row) for row in zip(*matrix)][::-1]
  raise ValueError("direction must be 'left' or 'right'")
````

## Problem 4

### Implementing a Library Management System

#### Description

You are required to design and implement a system for managing books and users in a library. The system should allow for the management of books (adding, deleting, searching by various criteria) and users (registration, deletion, searching), as well as tracking the history of interactions between them (issuing and returning books).

#### Tasks

1. **`Book` Class**:
   - Attributes: title, author, year of publication, ISBN, number of copies.
   - Methods: constructor, methods to get information about the book, method to change the number of copies (when issuing and returning books).

2. **`User` Class**:
   - Attributes: user name, library card number, list of borrowed books.
   - Methods: constructor, methods for user registration, methods for adding and removing books from the borrowed list.

3. **`Library` Class**:
   - Attributes: list of books, list of users, transaction history (who, when, which book was borrowed and returned).
   - Methods: constructor, methods for adding and deleting books and users, methods for issuing and returning books, searching for books and users by various criteria, method to display the transaction history.

#### Assignment

1. Implement the `Book`, `User`, and `Library` classes with the specified attributes and methods.
2. Create several books and users, and add them to the library system.
3. Implement scenarios for issuing books to users and their return.
4. Display the transaction history to show how books were issued and returned.


````python
from datetime import datetime

class Book:
    def __init__(self, title, author, year, isbn, copies=1):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = str(isbn)
        self.copies = int(copies)

    def get_info(self):
        return f"{self.title} — {self.author} ({self.year}), ISBN: {self.isbn}, copies: {self.copies}"

    def change_copies(self, delta):
        new_copies = self.copies + int(delta)
        if new_copies < 0:
            raise ValueError("Not enough copies to perform this operation.")
        self.copies = new_copies


class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = str(card_number)
        self.borrowed_books = []

    def borrow_book(self, isbn):
        isbn = str(isbn)
        self.borrowed_books.append(isbn)

    def return_book(self, isbn):
        isbn = str(isbn)
        if isbn not in self.borrowed_books:
            return False
        self.borrowed_books.remove(isbn)
        return True


class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.history = []

    def _find_book_by_isbn(self, isbn):
        isbn = str(isbn)
        for b in self.books:
            if b.isbn == isbn:
                return b
        return None

    def _find_user_by_card(self, card_number):
        card_number = str(card_number)
        for u in self.users:
            if u.card_number == card_number:
                return u
        return None

    def _log(self, message):
        stamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self.history.append(f"[{stamp}] {message}")

    def add_book(self, book):
        existing = self._find_book_by_isbn(book.isbn)
        if existing:
            existing.change_copies(book.copies)
            print(f"Added copies: {book.copies} -> {existing.get_info()}")
        else:
            self.books.append(book)
            print(f"Book added -> {book.get_info()}")

    def delete_book(self, isbn):
        book = self._find_book_by_isbn(isbn)
        if not book:
            print("Delete failed: book not found.")
            return False

        for u in self.users:
            if book.isbn in u.borrowed_books:
                print("Delete failed: this book is currently borrowed by a user.")
                return False

        self.books.remove(book)
        print(f"Book deleted: ISBN {isbn}")
        return True

    def search_books(self, title=None, author=None, year=None, isbn=None):
        res = []
        for b in self.books:
            if isbn is not None and b.isbn != str(isbn):
                continue
            if year is not None and b.year != year:
                continue
            if title is not None and title.lower() not in b.title.lower():
                continue
            if author is not None and author.lower() not in b.author.lower():
                continue
            res.append(b)
        return res

    def register_user(self, user):
        if self._find_user_by_card(user.card_number):
            print("Registration failed: card number already exists.")
            return False
        self.users.append(user)
        print(f"User registered: {user.name}, card: {user.card_number}")
        return True

    def delete_user(self, card_number):
        user = self._find_user_by_card(card_number)
        if not user:
            print("Delete failed: user not found.")
            return False
        if user.borrowed_books:
            print("Delete failed: user still has borrowed books.")
            return False
        self.users.remove(user)
        print(f"User deleted: card {card_number}")
        return True

    def search_users(self, name=None, card_number=None):
        res = []
        for u in self.users:
            if card_number is not None and u.card_number != str(card_number):
                continue
            if name is not None and name.lower() not in u.name.lower():
                continue
            res.append(u)
        return res

    def issue_book(self, card_number, isbn):
        user = self._find_user_by_card(card_number)
        book = self._find_book_by_isbn(isbn)

        if not user:
            print("Issue failed: user not found.")
            return False
        if not book:
            print("Issue failed: book not found.")
            return False
        if book.copies <= 0:
            print("Issue failed: no copies available.")
            return False

        book.change_copies(-1)
        user.borrow_book(book.isbn)
        self._log(f"{user.name} (card {user.card_number}) borrowed '{book.title}' (ISBN {book.isbn})")
        print(f"Issued -> {user.name} borrowed {book.title}. Remaining copies: {book.copies}")
        return True

    def return_book(self, card_number, isbn):
        user = self._find_user_by_card(card_number)
        book = self._find_book_by_isbn(isbn)

        if not user:
            print("Return failed: user not found.")
            return False
        if not book:
            print("Return failed: book not found.")
            return False

        if not user.return_book(book.isbn):
            print("Return failed: user did not borrow this book.")
            return False

        book.change_copies(+1)
        self._log(f"{user.name} (card {user.card_number}) returned '{book.title}' (ISBN {book.isbn})")
        print(f"Returned -> {user.name} returned {book.title}. Copies now: {book.copies}")
        return True

    def display_history(self):
        print("\n--- Transaction History ---")
        if not self.history:
            print("No transactions yet.")
        for line in self.history:
            print(line)


library = Library()

b1 = Book("1984", "George Orwell", 1949, "9780451524935", copies=2)
b2 = Book("The Hobbit", "J.R.R. Tolkien", 1937, "9780547928227", copies=1)
b3 = Book("Clean Code", "Robert C. Martin", 2008, "9780132350884", copies=1)

library.add_book(b1)
library.add_book(b2)
library.add_book(b3)

u1 = User("Alice Johnson", "U1001")
u2 = User("Bob Smith", "U1002")

library.register_user(u1)
library.register_user(u2)

library.issue_book("U1001", "9780451524935")
library.issue_book("U1002", "9780451524935")
library.issue_book("U1002", "9780547928227")
library.return_book("U1002", "9780451524935")

library.display_history()
````

## Problem 6*

Let
$$A = \sum_{i=1}^{10000} \frac{1}{i^2},\quad B=\sum_{i=10000}^{1} \frac{1}{i^2}.$$
Calculate the values of $A$ and $B$ and compare them. What do you observe? Explain why this happens. What is the best way to calculate the value of $\sum\limits_{i=1}^{10000} \dfrac{1}{i^2}$?

````python
A, B = 0, 0
for i in range(1, 10001):
    A += 1/(i**2)
for i in range(10000, 0, -1):
    B += 1/(i**2)

print(A)
print(B)
print(A-B)
````

Несовпадение сумм A и B связано с погрешностью вычислений с плавающей точкой. При разном порядке суммирования промежуточные округления дают разные результаты. Чтобы повысить точность, нужно суммировать числа от меньших к большим. Это уменьшает накопление ошибки
